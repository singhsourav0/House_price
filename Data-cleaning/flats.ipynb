Flats data cleaning

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
try:
    flats = pd.read_csv("flats.csv")
except FileNotFoundError as e:
    print(f"Error: {e}")
    print("Ensure the file exists at the specified path.")
```

```python
#for finding null values in columns
flats.isnull().sum() # 0 missing values in the dataset
```

```python
#for finding null values in rows
flats.notnull().sum(axis=1)
```

```python
# Set the minimum number of non-null values required
min_values_required = 10

# Detect rows with fewer non-null values
valid_rows = flats.notnull().sum(axis=1) >= min_values_required

# Filter out invalid rows
flats = flats[valid_rows]

print("\nCleaned DataFrame (rows with at least 10 non-null values):")
print(flats)
```

```python
#flats data having dublicates?
flats.duplicated().sum()
```

```python
flats.isnull().sum()
```

```python
#droping useless columns
flats.drop(columns= ['property_id','link'], inplace = True)
```

```python
flats.shape
```

```python
flats.columns
```

```python
flats.rename(columns={'area':'price_per_sqft'}, inplace= True)
flats.head(1)
```

```python
flats['society'].value_counts()
```

```python
flats['society'].value_counts().shape
```

```python
import re
flats['society'] = flats['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
flats['society'].value_counts().shape
```

```python
flats.sample(3)
```

```python
flats['price'].value_counts()
```

```python
flats[flats['price'] == 'Price on Request'].shape
```

```python
flats = flats[flats['price'] != 'Price on Request']
flats.shape
```

```python
#Clean the price 
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
flats["price"].isnull().sum() # 0
```

```python
flats = flats.dropna(subset=["price"])
flats["price"].isna().sum()
```

```python
flats["price"].unique()
```

```python
flats = flats[flats['price'] != 'price']
flats.shape
```

```python
flats['price'] = flats['price'].str.split(' ').apply(treat_price)
```

```python
flats.sample(2)
```

### Now Price columns are clean

```python
flats["price_per_sqft"].value_counts()
```

```python
flats['price_per_sqft'] = flats['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```

```python
flats.sample(1)
```

### Price_per_sqft is cleaned...

```python
flats["bedRoom"].value_counts()
```

```python
flats['bedRoom'].isna().sum()
```

```python
flats['bedRoom'] = flats['bedRoom'].str.split(' ').str.get(0).astype('int')
```

```python
flats['bedRoom'].unique()
```

```python
flats['bathroom'] = flats['bathroom'].str.split(' ').str.get(0).astype('int')
```

```python
flats['additionalRoom'].fillna('not avaliable', inplace = True)
flats.sample(5)
```

```python
flats['additionalRoom']= flats['additionalRoom'].str.lower()
flats.tail(2)
```

```python
flats[flats['floorNum'].isnull()]
```

```python
flats['floorNum'] = flats['floorNum'].str.split(' ').str.get(0).replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract(r'(\d+)')
```
