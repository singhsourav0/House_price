```python
# Set the minimum number of non-null values required
min_values_required = 10

# Detect rows with fewer non-null values
valid_rows = house.notnull().sum(axis=1) >= min_values_required

# Filter out invalid rows
house = house[valid_rows]

print("\nCleaned DataFrame (rows with at least 10 non-null values):")
print(house)
```

```python
import re
house['society'] = house['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
house['society'].value_counts().shape
```

```python
house['society'] = house['society'].str.replace('nan','independent')
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
house['price']= house["price"].str.split(" ").apply(treat_price)
house.sample(2)
```

```python
house['price_per_sqft'].value_counts()
```

```python
house['price_per_sqft']= house['price_per_sqft'].str.split("/").str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
house.head(2)
```
